Informações relevantes:

Os dados estão relacionados com campanhas de marketing direto de uma instituição bancária portuguesa. As campanhas de marketing foram baseadas em telefonemas. Muitas vezes, era necessário mais de um contato para o mesmo cliente, para acessar se o produto (depósito a prazo bancário) seria (ou não) subscrito.

Existem dois conjuntos de dados:

  1) bank-full.csv com todos os exemplos, ordenados por data (de maio de 2008 a novembro de 2010).

  2) bank.csv com 10% dos exemplos (4521), selecionados aleatoriamente de bank-full.csv.
O menor conjunto de dados é fornecido para testar algoritmos de aprendizado de máquina computacionalmente mais exigentes (por exemplo, SVM).

O objetivo da classificação é prever se o cliente irá subscrever um depósito a prazo (variável y).

Número de instâncias: 45211 para bank-full.csv (4521 para bank.csv)

Número de Atributos: 16 + atributo de saída.

Informações de atributo:

Para mais informações, leia [Moro et al., 2011].

Variáveis de entrada:

#dados do cliente do banco:

1 - idade (numérico)

2 - emprego : tipo de emprego(categórico:"administrador.","desconhecido","desempregado","gerente","empregada","empreendedor","aluno", "colarinho azul","autônomo","aposentado","técnico","serviços")

3 - marital : estado civil (categórico: "casado", "divorciado", "solteiro"; nota: "divorciado" significa divorciado ou viúvo)

4 - educação (categórica: "desconhecido", "secundário", "primário", "terciário")

5 - inadimplência: tem crédito inadimplente? (binário: "sim", "não")

6 - saldo: saldo médio anual, em euros (numérico)

7 - habitação: tem crédito habitação? (binário: "sim", "não")

8 - empréstimo: tem empréstimo pessoal? (binário: "sim", "não")

#relacionado com o último contato da campanha atual:

9 - contato: tipo de comunicação de contato (categórico: "desconhecido", "telefone", "celular")

10 - dia: último dia de contato do mês (numérico)

11 - mês: último mês de contato do ano (categórico: "jan", "fev", "mar", ..., "nov", "dec")

12 - duração: duração do último contato, em segundos (numérico)

#outros atributos:

13 - campanha: número de contatos realizados durante esta campanha e para este cliente (numérico, inclui último contato)

14 - pdays: número de dias que se passaram após o último contato com o cliente de uma campanha anterior (numérico, -1 significa que o cliente não foi contatado anteriormente)

15 - anterior: número de contatos realizados antes desta campanha e para este cliente (numérico)

16 - resultado: resultado da campanha de marketing anterior (categórica: "desconhecido", "outro", "fracasso", "sucesso")

Variável de saída (destino desejado):

17 - y - o cliente subscreveu um depósito a prazo? (binário: "sim", "não")

Valores de atributo ausentes: Nenhum

```python
#Importando as Bibliotecas

import numpy as np
import pandas as pd
from datatile.summary.df import DataFrameSummary
```

```python
# Carregando o dataset

df = pd.read_csv("bank.csv", sep = ';') 
```

Ao começar com um novo conjunto de dados, uma das maneiras comuns de obter mais informações sobre as várias colunas em um Pandas DataFrame é usando o método .info()and .describe().
O .info()método fornece informações sobre o número de nulos e o tipo de dados para cada coluna no DataFrame.

```python
# Verificando tipo de dados para cada coluna no DataFrame.

df.info()
```

O .describe()método fornece estatísticas resumidas para colunas numéricas no DataFrame.

```python
# Verificando as Estatísticas do dataframe
df.describe()
```

```python
# Agora vamos verificar os valores distintos em nosso dataframe
df.apply(lambda x: x.nunique())
```

Para encontrar a porcentagem de valores ausentes em cada coluna

```python
df.isnull().sum() * 100 / len(df)  # Observe que não há nenhum valor null em nenhuma coluna
```
